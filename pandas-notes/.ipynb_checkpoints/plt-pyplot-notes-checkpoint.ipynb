#### matplotlib.pyplot is a collection of command style functions that make matplotlib work like MATLAB.  
Each **pyplot** function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure,  plots some lines in a plotting area, decorates the plot with labels, etc. 
- Source: matplotlib.org

# 1. 



```python
import matplotlib.pyplot as plt 
plt.plot([1,2,3,4])
plt.ylabel('some numbers')
plt.show()
```

'''x-axis is generated atuomatically. matplotlib assistes when a [single array is provided] that it is a sequence of y values.'''

# 2

```python
plt.plot([1,2,3,4], [8,10,12,14])
plt.axis([0,6,0,20])
plt.show()
```

* same but now with an optional argument for color and line of the plot. green circles

```python
plt.plot([1,2,3,4], [8,10,12,14], 'go')
plt.axis([0,6,0,20]) #axis command takes [xmin,xmax,ymin,ymax]
plt.show()
```

* using numpy arrays as a sample list, arrange several lines of numbers with different format styles: 

```python
import numpy as np
# evenly sampled time at 200ms intervals
t = np.arange(0., 5.,0.2)
# red dashes, blue squares, green circles
plt.plot(t, t, 'r--', t,t*2, 'bs', t, t*3, 'go') 
plt.show()
```
